

### **How the Human Ear Works (From an Audio Processing Viewpoint)**
The ear processes sound in three parts: **outer ear, middle ear, and inner ear**.

1. **Outer Ear:**
   - Soundwaves enter the ear and hit the **eardrum** (a thin membrane).
   - The eardrum vibrates based on the loudness and pitch of the sound.

2. **Middle Ear:**
   - Vibrations from the eardrum are passed to the **inner ear**.
   - It acts like a filter, allowing certain frequencies to pass through.

3. **Inner Ear:**
   - The inner ear contains the **cochlea**, a coiled structure with a **basilar membrane** inside.
   - The basilar membrane is like a built-in **spectrum analyzer**, sensitive to specific frequencies at different points. This helps us perceive different sounds.

---

### **Modeling the Ear for Audio Processing**
The way the ear processes sound can be represented using **digital filters**:
- The **basilar membrane** is divided into many small sections, each acting as a filter.
- There are **128 filters** in total, but processing sound through all of them one by one is too slow for real-time applications.
- To fix this, engineers use **32 parallel filters** (called **critical bands**) to speed up the process.

---

### **Human Auditory Masking**
The human ear has limitations:
- **Masking** occurs when one sound overpowers another, making the quieter sound inaudible.

#### Types of Masking:
1. **Frequency Masking (Simultaneous Masking):**
   - Example: If you’re talking to someone and a loud car passes by, your conversation might be drowned out by the car noise.
   
2. **Temporal Masking (Non-Simultaneous Masking):**
   - Example: If a loud sound happens shortly before or after a quieter sound, the quieter sound might still be masked.

These masking effects are used to compress audio by removing inaudible parts.

---

### **How Does Audio Compression Work?**
Compression reduces the size of audio files by:
1. Splitting the frequency range into bands (e.g., **Bark bands**, which are logarithmically organized critical bands).
2. Removing inaudible parts based on masking.
3. Encoding only the audible parts, resulting in smaller file sizes.

This method is known as **lossy compression** because the removed information cannot be restored.

---

### **Popular Audio Codecs and How They Work**
1. **MP3 (MPEG-1 Audio Layer 3):**
   - Breaks audio into frequency bands using **filters**.
   - Converts signals from the time domain to the frequency domain using **MDCT (Modified Discrete Cosine Transform)**.
   - Removes inaudible sounds based on psychoacoustic models.

2. **AAC (Advanced Audio Coding):**
   - An improved version of MP3 that offers better compression.

3. **WMA (Windows Media Audio):**
   - Microsoft’s proprietary codec, often used for paid music.

4. **Vorbis:**
   - A free codec supporting multi-channel compression, often used in open-source projects.

5. **FLAC (Free Lossless Audio Codec):**
   - A lossless codec that preserves all audio data but results in larger file sizes.

6. **AC-3:**
   - Designed by Dolby for multi-channel audio (e.g., surround sound).

The techniques discussed above are widely used in:
